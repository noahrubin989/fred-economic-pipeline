## Sandbox Area

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("../data/economic_data.db")
df = pd.read_sql("SELECT * FROM observations", conn)
```

### Basic Cleaning

```python
df["date"] = pd.to_datetime(df["date"])
df["value"] = pd.to_numeric(df["value"], errors="coerce")
print(df.dtypes)
```

### Date Exploration

```python
df["series_id"].value_counts()
df.groupby("series_id")["date"].agg(["min", "max"])
```
